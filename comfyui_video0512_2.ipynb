```python
# ==========================================
# Cell 3: 终极模型多线程极速同步 (精准路径修正版)
# ==========================================
import os
import shutil
from google.colab import userdata
from huggingface_hub import hf_hub_download
from concurrent.futures import ThreadPoolExecutor

os.environ["HF_HUB_ENABLE_HF_TRANSFER"] = "1"
try:
    os.environ["HF_TOKEN"] = userdata.get('HF_TOKEN')
except:
    raise Exception("❌ 请在左侧 '🔑 Secrets' 中添加 'HF_TOKEN'")

# 这一次，repo_id 和 filename 被精准映射到了各自真实的 HF 仓库
downloads = [
    # --- 1. 之前已成功的基础模型与 LoRA ---
   # {"repo_id": "Lightricks/LTX-2.3", "filename": "ltx-2.3-22b-distilled-lora-384-1.1.safetensors", "local_dir": "/content/ComfyUI/models/loras"},
    #{"repo_id": "Lightricks/LTX-2.3", "filename": "ltx-2.3-spatial-upscaler-x2-1.1.safetensors", "local_dir": "/content/ComfyUI/models/latent_upscale_models"},
   # {"repo_id": "Lightricks/LTX-2.3-22b-IC-LoRA-Union-Control", "filename": "ltx-2.3-22b-ic-lora-union-control-ref0.5.safetensors", "local_dir": "/content/ComfyUI/models/loras"},
   # {"repo_id": "Lightricks/LTX-2.3", "filename": "ltx-2.3-22b-distilled-1.1.safetensors", "local_dir": "/content/ComfyUI/models/checkpoints"},

    # --- 2. VAE 模型 (Kijai 大佬的拆分仓库) ---
    {"repo_id": "Kijai/LTX2.3_comfy", "filename": "vae/LTX23_audio_vae_bf16.safetensors", "local_dir": "/content/ComfyUI/models/vae"},
    {"repo_id": "Kijai/LTX2.3_comfy", "filename": "vae/LTX23_video_vae_bf16.safetensors", "local_dir": "/content/ComfyUI/models/vae"},
    {"repo_id": "Kijai/LTX2.3_comfy", "filename": "vae/taeltx2_3.safetensors", "local_dir": "/content/ComfyUI/models/vae"},

    # --- 3. Text Encoders / CLIP ---
    # Gemma 在 Comfy-Org 的 LTX-2 (注意不是 2.3) 仓库里
    {"repo_id": "Comfy-Org/ltx-2", "filename": "split_files/text_encoders/gemma_3_12B_it_fp4_mixed.safetensors", "local_dir": "/content/ComfyUI/models/text_encoders"},
    # 投影层在 Kijai 的仓库里
    {"repo_id": "Kijai/LTX2.3_comfy", "filename": "text_encoders/ltx-2.3_text_projection_bf16.safetensors", "local_dir": "/content/ComfyUI/models/text_encoders"},

    # --- 4. UNET / Diffusion Models ---
    # GGUF 低显存版实际上由 vantagewithai / unsloth 维护
    #{"repo_id": "vantagewithai/LTX-2.3-GGUF", "filename": "dev/ltx-2-3-22b-dev-Q4_K_M.gguf", "local_dir": "/content/ComfyUI/models/unet"},
    # FP8 缩放版放在 Kijai 仓库的 diffusion_models 目录下
    #{"repo_id": "Kijai/LTX2.3_comfy", "filename": "diffusion_models/ltx-2.3-22b-dev_transformer_only_fp8_scaled.safetensors", "local_dir": "/content/ComfyUI/models/unet"},

    # --- 5. 镜头控制及动态 LoRA 组 ---
    # 动态控制 LoRA 在 Kijai 仓库
    {"repo_id": "Kijai/LTX2.3_comfy", "filename": "loras/ltx-2.3-22b-distilled-lora-dynamic_fro09_avg_rank_105_bf16.safetensors", "local_dir": "/content/ComfyUI/models/loras"},

    # 运镜 LoRA 官方为每个方向都单独建立了一个独立仓库！
    {"repo_id": "Lightricks/LTX-2-19b-LoRA-Camera-Control-Dolly-Left", "filename": "ltx-2-19b-lora-camera-control-dolly-left.safetensors", "local_dir": "/content/ComfyUI/models/loras"},
    {"repo_id": "Lightricks/LTX-2-19b-LoRA-Camera-Control-Dolly-Right", "filename": "ltx-2-19b-lora-camera-control-dolly-right.safetensors", "local_dir": "/content/ComfyUI/models/loras"},
    {"repo_id": "Lightricks/LTX-2-19b-LoRA-Camera-Control-Dolly-In", "filename": "ltx-2-19b-lora-camera-control-dolly-in.safetensors", "local_dir": "/content/ComfyUI/models/loras"},
    {"repo_id": "Lightricks/LTX-2-19b-LoRA-Camera-Control-Dolly-Out", "filename": "ltx-2-19b-lora-camera-control-dolly-out.safetensors", "local_dir": "/content/ComfyUI/models/loras"},
    {"repo_id": "Lightricks/LTX-2-19b-LoRA-Camera-Control-Jib-Up", "filename": "ltx-2-19b-lora-camera-control-jib-up.safetensors", "local_dir": "/content/ComfyUI/models/loras"},
    {"repo_id": "Lightricks/LTX-2-19b-LoRA-Camera-Control-Jib-Down", "filename": "ltx-2-19b-lora-camera-control-jib-down.safetensors", "local_dir": "/content/ComfyUI/models/loras"}
]

def download_and_flatten_model(task):
    # 提取纯文件名用于显示和最终路径
    display_name = task["filename"].split("/")[-1]
    target_dir = task["local_dir"]
    final_path = os.path.join(target_dir, display_name)

    try:
        # 下载到 HF 本地缓存 (不保留复杂的子文件夹结构)
        cached_path = hf_hub_download(repo_id=task["repo_id"], filename=task["filename"], repo_type="model")

        # 确保目标文件夹存在
        os.makedirs(target_dir, exist_ok=True)

        # 使用硬链接(或复制)直接把文件放在 ComfyUI 需要的最外层目录，避免节点找不到子文件夹内的文件
        if not os.path.exists(final_path):
            try:
                os.link(cached_path, final_path) # 首选硬链接，省空间
            except OSError:
                shutil.copy2(cached_path, final_path) # 跨盘符则复制

        print(f"✅ 成功加载: {display_name}")
    except Exception as e:
        print(f"❌ 失败 {display_name}: {e}")

print(f"🚀 开始精准校对并下载 {len(downloads)} 个模型...")

# 保持 6 线程
with ThreadPoolExecutor(max_workers=6) as executor:
    executor.map(download_and_flatten_model, downloads)

print("\n🎉 模型并发同步全部完成！请在 ComfyUI 界面点击 Refresh 刷新模型列表。")
```

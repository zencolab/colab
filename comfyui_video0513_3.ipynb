```python
# ==========================================
# Cell 4: еҗҜеҠЁ FRP е’Ң ComfyUI (йҮҚеҗҜз•Ңйқўж—¶иҝҗиЎҢ),дёҚжіЁеҶҢпјҢеҗҜеҠЁеҝ«пјҒи®ҝй—®еҹҹеҗҚ
# ==========================================
import subprocess
import threading
import os
import configparser

import time

# --- ж–°еўһпјҡйҳІж–ӯејҖдҝқжҙ»жңәеҲ¶ ---
def keep_alive():
    while True:
        time.sleep(300)  # жҜҸ 5 еҲҶй’ҹ
        print("\n[Keep-Alive] дҝқжҢҒ Colab иҝһжҺҘжҙ»и·ғдёӯ...")

print("вҸі жӯЈеңЁеҗҜеҠЁйҳІж–ӯејҖеҗҺеҸ°дҝқжҙ»зәҝзЁӢ...")
threading.Thread(target=keep_alive, daemon=True).start()
# --------------------------

# 1. еңЁеҗҺеҸ°ж–°зәҝзЁӢеҗҜеҠЁ FRP
def start_frpc():
    subprocess.run(["/content/frp_0.56.0_linux_amd64/frpc", "-c", "/content/frp_0.56.0_linux_amd64/frpc.toml"])

print("вҸі жӯЈеңЁеҗҺеҸ°е”Өиө· FRP з©ҝйҖҸжңҚеҠЎ...")
threading.Thread(target=start_frpc, daemon=True).start()

print("============================================================")
print("вң… FRP з©ҝйҖҸе·Іе°ұз»ӘпјҒ")
# иҝҷйҮҢе·Ідҝ®ж”№дёәдҪ зҡ„жҢҮе®ҡеҹҹеҗҚпјҲдҝқз•ҷдәҶ 8080 з«ҜеҸЈпјҢеҰӮжһңдҪ зҡ„жңҚеҠЎз«Ҝз”Ё Nginx еҒҡдәҶ 80 з«ҜеҸЈеҸҚд»ЈпјҢеҸҜд»ҘжҠҠ :8080 еҲ жҺүпјү
print("рҹ‘ү еҗҜеҠЁе®ҢжҲҗеҗҺпјҢиҜ·йҖҡиҝҮжөҸи§ҲеҷЁи®ҝй—®: http://cjp.usdream.dpdns.org:8090")
print("============================================================\n")

# ---------------------------------------------------------
# ж–°еўһпјҡеҪ»еә•жӢҰжҲӘ ComfyUI-Manager еҗҜеҠЁж—¶зҡ„иҒ”зҪ‘ Fetch иЎҢдёә
# ---------------------------------------------------------
print("вҸі жӯЈеңЁй…ҚзҪ® Manager зҪ‘з»ңжЁЎејҸд»Ҙи·іиҝҮ Fetch...")
# е…је®№жүҖжңүзүҲжң¬зҡ„й…ҚзҪ®ж–Үд»¶и·Ҝеҫ„ (зү№еҲ«жҳҜжңҖж–°зҡ„ __manager и·Ҝеҫ„)
manager_config_paths = [
    "/content/ComfyUI/user/__manager/config.ini",               # жңҖж–°зүҲ V3.39+ й…ҚзҪ®и·Ҝеҫ„
    "/content/ComfyUI/user/default/ComfyUI-Manager/config.ini", # иҫғж–°зүҲй…ҚзҪ®и·Ҝеҫ„
    "/content/ComfyUI/custom_nodes/ComfyUI-Manager/config.ini"  # иҖҒж—§зүҲй…ҚзҪ®и·Ҝеҫ„
]

for config_path in manager_config_paths:
    os.makedirs(os.path.dirname(config_path), exist_ok=True)
    config = configparser.ConfigParser()

    if os.path.exists(config_path):
        config.read(config_path)

    if 'default' not in config:
        config['default'] = {}

    # е°ҶзҪ‘з»ңжЁЎејҸж”№дёә private
    config['default']['network_mode'] = 'private'

    with open(config_path, 'w') as f:
        config.write(f)

print("вң… е·ІжҲҗеҠҹеҲҮж–ӯ Fetch ComfyRegistry Data жөҒзЁӢпјҒ\n")
# ---------------------------------------------------------

# 2. еҗҜеҠЁ ComfyUI дё»зЁӢеәҸ
print("вҸі жӯЈеңЁеҗҜеҠЁ ComfyUI дё»иҝӣзЁӢ...\n")
%cd /content/ComfyUI
!python main.py --dont-print-server
```

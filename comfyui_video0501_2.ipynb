```python
# ==========================================
# Cell 4: з»ҲжһҒжЁЎеһӢеӨҡзәҝзЁӢжһҒйҖҹеҗҢжӯҘ (жһҒеӨ§еҠ еҝ«дёӢиҪҪ)
# ==========================================
import os
from google.colab import userdata
from huggingface_hub import hf_hub_download
from concurrent.futures import ThreadPoolExecutor

os.environ["HF_HUB_ENABLE_HF_TRANSFER"] = "1"
try:
    os.environ["HF_TOKEN"] = userdata.get('HF_TOKEN')
except:
    raise Exception("вқҢ иҜ·еңЁе·Ұдҫ§ 'рҹ”‘ Secrets' дёӯж·»еҠ  'HF_TOKEN'")

downloads = [
    # --- еҺҹжңүзҡ„жЁЎеһӢ ---
    #{"repo_id": "Lightricks/LTX-2.3-fp8", "filename": "ltx-2.3-22b-dev-fp8.safetensors", "local_dir": "/content/ComfyUI/models/checkpoints"},
    {"repo_id": "Lightricks/LTX-2.3", "filename": "ltx-2.3-22b-distilled-lora-384.safetensors", "local_dir": "/content/ComfyUI/models/loras"},
    #{"repo_id": "Lightricks/LTX-2.3", "filename": "ltx-2.3-spatial-upscaler-x2-1.1.safetensors", "local_dir": "/content/ComfyUI/models/latent_upscale_models"}, # жӯӨжЁЎеһӢеҺҹд»Јз Ғдёӯе·ІеҢ…еҗ«
    #{"repo_id": "Comfy-Org/ltx-2", "filename": "split_files/text_encoders/gemma_3_12B_it_fp4_mixed.safetensors", "local_dir": "/content/ComfyUI/models/text_encoders"},
    #{"repo_id": "Comfy-Org/ltx-2.3", "filename": "split_files/loras/ltx-2.3-id-lora-talkvid-3k.safetensors", "local_dir": "/content/ComfyUI/models/loras"},
    {"repo_id": "Lightricks/LTX-2.3-22b-IC-LoRA-Union-Control", "filename": "ltx-2.3-22b-ic-lora-union-control-ref0.5.safetensors", "local_dir": "/content/ComfyUI/models/loras"},

    # --- жң¬ж¬Ўж–°еўһзҡ„жЁЎеһӢ ---
    # еҹәзЎҖжЁЎеһӢ (ж”ҫе…Ҙ checkpoints ж–Үд»¶еӨ№)
    {"repo_id": "Lightricks/LTX-2.3", "filename": "ltx-2.3-22b-dev.safetensors", "local_dir": "/content/ComfyUI/models/checkpoints"},
    # ж—¶й—ҙж”ҫеӨ§жЁЎеһӢ (ж”ҫе…Ҙ latent_upscale_models ж–Үд»¶еӨ№)
    #{"repo_id": "Lightricks/LTX-2.3", "filename": "ltx-2.3-temporal-upscaler-x2-1.0.safetensors", "local_dir": "/content/ComfyUI/models/latent_upscale_models"}
]

def download_model(task):
    try:
        hf_hub_download(repo_id=task["repo_id"], filename=task["filename"], local_dir=task["local_dir"], repo_type="model")
        print(f"вң… е®ҢжҲҗ: {task['filename']}")
    except Exception as e:
        print(f"вқҢ еӨұиҙҘ {task['filename']}: {e}")

print(f"рҹҡҖ ејҖе§ӢеҗҜеҠЁеӨҡзәҝзЁӢе№¶еҸ‘дёӢиҪҪ {len(downloads)} дёӘжЁЎеһӢ...")
# еҸҜд»Ҙж №жҚ®йңҖиҰҒйҖӮеәҰи°ғй«ҳзәҝзЁӢж•°пјҢдҪҶColabе…Қиҙ№зүҲе»әи®®дҝқжҢҒеңЁ 6-8 е·ҰеҸід»Ҙе…Қиў«йҷҗжөҒ
with ThreadPoolExecutor(max_workers=6) as executor:
    executor.map(download_model, downloads)

print("\nрҹҺү жЁЎеһӢе№¶еҸ‘еҗҢжӯҘе…ЁйғЁе®ҢжҲҗпјҒ")
```

```python
# ==========================================
# Cell 5: FRP еҶ…зҪ‘з©ҝйҖҸй…ҚзҪ® (жөҒејҸи§ЈеҺӢжһҒйҖҹйғЁзҪІ)
# ==========================================
import os
from google.colab import userdata

try:
    VPS_IP = userdata.get('VPS_IP')
    FRP_TOKEN = userdata.get('FRP_TOKEN')
except:
    raise Exception("вқҢ иҜ·жЈҖжҹҘ 'VPS_IP' е’Ң 'FRP_TOKEN' еҜҶй’ҘжҳҜеҗҰй…ҚзҪ®пјҒ")

# жһҒйҖҹжөҒејҸдёӢиҪҪи§ЈеҺӢ
if not os.path.exists("/content/frp_0.56.0_linux_amd64"):
    !wget -qO- https://github.com/fatedier/frp/releases/download/v0.56.0/frp_0.56.0_linux_amd64.tar.gz | tar -xz -C /content

frpc_conf = f"""
serverAddr = "{VPS_IP}"
serverPort = 7000
auth.token = "{FRP_TOKEN}"

[[proxies]]
name = "comfyui_web"
type = "tcp"
localIP = "127.0.0.1"
localPort = 8188
remotePort = 8080
"""
with open("/content/frp_0.56.0_linux_amd64/frpc.toml", "w") as f:
    f.write(frpc_conf.strip())

print("вң… FRP жһҒйҖҹйғЁзҪІе№¶й…ҚзҪ®е®ҢжҜ•пјҒ")
```

```python
# ==========================================
# еӣӣгҖҒеҗҜеҠЁ FRP е’Ң ComfyUI (йҮҚеҗҜз•Ңйқўж—¶иҝҗиЎҢ),дёҚжіЁеҶҢпјҢеҗҜеҠЁеҝ«пјҒи®ҝй—®еҹҹеҗҚ
# ==========================================
import subprocess
import threading
import os
import configparser

import time

# --- ж–°еўһпјҡйҳІж–ӯејҖдҝқжҙ»жңәеҲ¶ ---
def keep_alive():
    while True:
        time.sleep(300)  # жҜҸ 5 еҲҶй’ҹ
        print("\n[Keep-Alive] дҝқжҢҒ Colab иҝһжҺҘжҙ»и·ғдёӯ...")

print("вҸі жӯЈеңЁеҗҜеҠЁйҳІж–ӯејҖеҗҺеҸ°дҝқжҙ»зәҝзЁӢ...")
threading.Thread(target=keep_alive, daemon=True).start()
# --------------------------

# 1. еңЁеҗҺеҸ°ж–°зәҝзЁӢеҗҜеҠЁ FRP
def start_frpc():
    subprocess.run(["/content/frp_0.56.0_linux_amd64/frpc", "-c", "/content/frp_0.56.0_linux_amd64/frpc.toml"])

print("вҸі жӯЈеңЁеҗҺеҸ°е”Өиө· FRP з©ҝйҖҸжңҚеҠЎ...")
threading.Thread(target=start_frpc, daemon=True).start()

print("============================================================")
print("вң… FRP з©ҝйҖҸе·Іе°ұз»ӘпјҒ")
# иҝҷйҮҢе·Ідҝ®ж”№дёәдҪ зҡ„жҢҮе®ҡеҹҹеҗҚпјҲдҝқз•ҷдәҶ 8080 з«ҜеҸЈпјҢеҰӮжһңдҪ зҡ„жңҚеҠЎз«Ҝз”Ё Nginx еҒҡдәҶ 80 з«ҜеҸЈеҸҚд»ЈпјҢеҸҜд»ҘжҠҠ :8080 еҲ жҺүпјү
print("рҹ‘ү еҗҜеҠЁе®ҢжҲҗеҗҺпјҢиҜ·йҖҡиҝҮжөҸи§ҲеҷЁи®ҝй—®: http://cjp.usdream.dpdns.org:8080")
print("============================================================\n")

# ---------------------------------------------------------
# ж–°еўһпјҡеҪ»еә•жӢҰжҲӘ ComfyUI-Manager еҗҜеҠЁж—¶зҡ„иҒ”зҪ‘ Fetch иЎҢдёә
# ---------------------------------------------------------
print("вҸі жӯЈеңЁй…ҚзҪ® Manager зҪ‘з»ңжЁЎејҸд»Ҙи·іиҝҮ Fetch...")
# е…је®№жүҖжңүзүҲжң¬зҡ„й…ҚзҪ®ж–Үд»¶и·Ҝеҫ„ (зү№еҲ«жҳҜжңҖж–°зҡ„ __manager и·Ҝеҫ„)
manager_config_paths = [
    "/content/ComfyUI/user/__manager/config.ini",               # жңҖж–°зүҲ V3.39+ й…ҚзҪ®и·Ҝеҫ„
    "/content/ComfyUI/user/default/ComfyUI-Manager/config.ini", # иҫғж–°зүҲй…ҚзҪ®и·Ҝеҫ„
    "/content/ComfyUI/custom_nodes/ComfyUI-Manager/config.ini"  # иҖҒж—§зүҲй…ҚзҪ®и·Ҝеҫ„
]

for config_path in manager_config_paths:
    os.makedirs(os.path.dirname(config_path), exist_ok=True)
    config = configparser.ConfigParser()

    if os.path.exists(config_path):
        config.read(config_path)

    if 'default' not in config:
        config['default'] = {}

    # е°ҶзҪ‘з»ңжЁЎејҸж”№дёә private
    config['default']['network_mode'] = 'private'

    with open(config_path, 'w') as f:
        config.write(f)

print("вң… е·ІжҲҗеҠҹеҲҮж–ӯ Fetch ComfyRegistry Data жөҒзЁӢпјҒ\n")
# ---------------------------------------------------------

# 2. еҗҜеҠЁ ComfyUI дё»зЁӢеәҸ
print("вҸі жӯЈеңЁеҗҜеҠЁ ComfyUI дё»иҝӣзЁӢ...\n")
%cd /content/ComfyUI
!python main.py --dont-print-server
```

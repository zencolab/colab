```python
# ==========================================
# еӣӣгҖҒеҗҜеҠЁ FRP е’Ң ComfyUI (йҮҚеҗҜз•Ңйқўж—¶иҝҗиЎҢ)
# ==========================================
import subprocess
import threading
from google.colab import userdata

# иҺ·еҸ– IP д»…дёәдәҶеңЁжҺ§еҲ¶еҸ°еҸӢеҘҪең°жү“еҚ°и®ҝй—®й“ҫжҺҘ
try:
    VPS_IP = userdata.get('VPS_IP')
except:
    VPS_IP = "дҪ зҡ„е…¬зҪ‘IP"

# 1. еңЁеҗҺеҸ°ж–°зәҝзЁӢеҗҜеҠЁ FRP (иҜ»еҸ–еҲҡжүҚз”ҹжҲҗзҡ„й…ҚзҪ®ж–Үд»¶)
def start_frpc():
    subprocess.run(["/content/frp_0.56.0_linux_amd64/frpc", "-c", "/content/frp_0.56.0_linux_amd64/frpc.toml"])

print("вҸі жӯЈеңЁеҗҺеҸ°е”Өиө· FRP з©ҝйҖҸжңҚеҠЎ...")
threading.Thread(target=start_frpc, daemon=True).start()

print("============================================================")
print("вң… FRP з©ҝйҖҸе·Іе°ұз»ӘпјҒ")
print(f"рҹ‘ү еҗҜеҠЁе®ҢжҲҗеҗҺпјҢиҜ·йҖҡиҝҮжөҸи§ҲеҷЁи®ҝй—®: http://{VPS_IP}:8080")
print("============================================================\n")

# 2. еҗҜеҠЁ ComfyUI дё»зЁӢеәҸ
print("вҸі жӯЈеңЁеҗҜеҠЁ ComfyUI дё»иҝӣзЁӢ...\n")
%cd /content/ComfyUI
!python main.py --dont-print-server
```
